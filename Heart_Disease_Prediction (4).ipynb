```python
X=df.drop('TenYearCHD', axis=1)
```

```python
X
```

```python
Y=df['TenYearCHD']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=50)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train_scaler=scaler.fit_transform(x_train)
x_test_scaler=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,classification_report,roc_auc_score
```

```python
lr=LogisticRegression(max_iter=500)
param_grid={
    'penalty':['l1','l2'],
    'C':np.logspace(-3,3,10),
    'solver':['liblinear']
}
```

```python
grid=GridSearchCV(estimator=lr,param_grid=param_grid,scoring='accuracy',cv=5,n_jobs=-1)
```

```python
grid.fit(x_train_scaler,y_train)
```

```python
best_model=grid.best_estimator_
best_params=grid.best_params_
```

```python
y_pred=best_model.predict(x_test_scaler)
```

```python
acc=accuracy_score(y_pred,y_test)
pre=precision_score(y_pred,y_test,average='weighted')
rec=recall_score(y_pred,y_test,average='weighted')
f1=f1_score(y_pred,y_test,average='weighted')
roc_auc=roc_auc_score(y_pred,y_test)
```

```python
print("Accuracy:", acc)
print("Precision:", pre)
print("Recall:", rec)
print("f1_score:", f1)
print("Roc-Auc:",roc_auc)
```

```python
print("confusion matrix\n", confusion_matrix(y_pred,y_test))
print("classification report\n", classification_report(y_pred,y_test))
```
